```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def walker():
    global x,y
    eleccion = random.randint(0,3)
    if eleccion == 0:
        x+=1
    elif eleccion ==1:
        x-=1
    elif eleccion ==2:
        y+=1
    else:
        y-=1
        return [x,y]
    
x=0 #inicia el caminador
y=0

Pos = np.zeros((2,10000))

for i in np.arange(0,len(Pos[0]),1):
    Pos[:,i] = walker()
    
plt.plot(Pos[0,:],Pos[1,:],'bo-')
plt.plot(Pos[0,-1],Pos[1,-1],'ro-')
```

```python
%%timeit
#elements = np.array [1,0],[-1,0],[0,1],[1,-1]

inicio = np.array ([[0],[0]])

elements = np.array ([[1,-1,0,0],[0,0,1,-1]])
probabilites = [0.4, 0.2, 0.2, 0.2]

pasos = elements[:,np.random.choice(elements.shape[1], 10000, probabilites)]
pasos = np.concatenate( (inicio,pasos),axis=1)

Pos = np.cumsum (pasos ,axis=1 )
plt.plot(Pos[0,:],Pos[1,:],'bo-')
```

```python
# Numero de puntos tirados aleatoriamente
N= 1000
f= lambda x: np.sin(x)
a,b = 0,np.pi
Altura= 1

x = np.random.uniform(a,b,N)
y = np.random.uniform(a,Altura,N)
X = np.arange(a,b,0.01)
fx = f(x)

plt.plot(x[y>fx],y[y>fx],"go")
plt.plot(x[y<fx],y[y<fx],"ro")
plt.plot(X,f(X))


N_0= len(x[y<fx])
Base = b-a
yest =N_0/N*(Base*Altura)
print (yest)
```
